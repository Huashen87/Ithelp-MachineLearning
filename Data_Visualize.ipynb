# Data Visualize

### (1) Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
import requests
```

### (2) Import Data Set

```python
url = 'https://github.com/1010code/iris-dnn-tensorflow/raw/master/data/Iris.csv'
s = requests.get(url).content
df_train = pd.read_csv(io.StringIO(s.decode('utf-8')))
df_train = df_train.drop(labels=['Id'],axis=1)
df_train
```

### (3) Map Label Value

```python
classMap = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
df_train['class'] = df_train['Species'].map(classMap)
df_train
```

### (4) Exploratory Data Analysis (EDA)

```python
df_train.hist(alpha=0.6, figsize=(6,6))
plt.tight_layout()
plt.show()
```

### (5) Kernel Density Estimation (KDE)

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df_train, figsize=(10, 10), diagonal='kde')
'Kernel Density Estimation'
```

```python
sns.pairplot(df_train, hue='Species', height=2, diag_kind='kde')
```

### (6) Correlation Map

```python
corr = df_train.drop(labels=['Species'], axis=1).corr()
plt.figure(figsize=(8,8))
sns.heatmap(corr, square=True, annot=True, cmap='RdBu_r')
```

```python
sns.lmplot('SepalLengthCm', 'SepalWidthCm', hue='Species', data=df_train, fit_reg=False)
```

```python
sns.lmplot('PetalLengthCm', 'PetalWidthCm', hue='Species', data=df_train, fit_reg=False)
```

### (7) Box Diagram

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(8,5), sharey=True)
axes[0].boxplot(df_train['SepalLengthCm'], showmeans=True)
axes[0].set_title('SepalLengthCm')

axes[1].boxplot(df_train['SepalWidthCm'], showmeans=True)
axes[1].set_title('SepalWidthCm')

axes[2].boxplot(df_train['PetalLengthCm'], showmeans=True)
axes[2].set_title('PetalLengthCm')

axes[3].boxplot(df_train['PetalWidthCm'], showmeans=True)
axes[3].set_title('PetalWidthCm')
```

### (8) Distribution Diagram

```python
fig, axes = plt.subplots(nrows=1, ncols=4)
fig.set_size_inches(15,4)
sns.distplot(df_train['SepalLengthCm'], ax=axes[0])
sns.distplot(df_train['SepalWidthCm'], ax=axes[1])
sns.distplot(df_train['PetalLengthCm'], ax=axes[2])
sns.distplot(df_train['PetalWidthCm'], ax=axes[3])

axes[0].set(xlabel='SepalLengthCm', ylabel='', title='Distribution of SepalLengthCm')
axes[1].set(xlabel='SepalWidthCm', ylabel='', title='Distribution of SepalWidthCm')
axes[2].set(xlabel='PetalLengthCm', ylabel='', title='Distribution of PetalLengthCm')
axes[3].set(xlabel='PetalWidthCm', ylabel='', title='Distribution of PetalWidthCm')
```
